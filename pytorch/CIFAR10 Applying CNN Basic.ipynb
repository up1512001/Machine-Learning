```python
import numpy as np

def split_indices(n,val_pct=0.1,seed=99):
    # determine size of validation set
    n_val = int(val_pct*n)
    # set random seed for reproducibility
    np.random.seed(seed)
    # create random permutation of 0 to n-1
    idxs = np.random.permutation(n)
    # pick last n_val indices for validation set
    return idxs[n_val:],idxs[:n_val]
```

```python
val_pct = 0.2
rand_seed = 42

train_indices,val_indices = split_indices(len(dataset),val_pct,rand_seed)
print(len(train_indices),len(val_indices))
print('Sample Validation set :',val_indices[:10])
```

```python
# training sampler and data loader
train_sampler = SubsetRandomSampler(train_indices)
train_dl = DataLoader(dataset,batch_size,sampler=train_sampler)

# validation sampler and data loader
val_sampler = SubsetRandomSampler(val_indices)
val_dl = DataLoader(dataset,batch_size,sampler=val_sampler)
```

```python
# creating model

model = nn.Sequential(
        nn.Conv2d(3,16,kernel_size=3,stride=1,padding=1),
        nn.ReLU(),
        nn.MaxPool2d(2,2), # outputs : batch_size X 16 X 16 X 16
    
        nn.Conv2d(16,16,kernel_size=3,stride=1,padding=1),
        nn.ReLU(),
        nn.MaxPool2d(2,2), # outputs : bs X 16 X 8 X 8
    
        nn.Conv2d(16,16,kernel_size=3,stride=1,padding=1),
        nn.ReLU(),
        nn.MaxPool2d(2,2), # outputs : bs X 16 X 4 X 4
    
        nn.Conv2d(16,16,kernel_size=3,stride=1,padding=1),
        nn.ReLU(),
        nn.MaxPool2d(2,2), # outputs : bs X 16 X 2 X 2
    
        nn.Conv2d(16,16,kernel_size=3,stride=1,padding=1),
        nn.ReLU(),
        nn.MaxPool2d(2,2), # outputs : bs X 16 X 1 X 1
    
        nn.Flatten(), # outputs : bs X 16
        nn.Linear(16,10) # outputs : bs X 10
)
```

```python
def get_default_device():
    '''pick GUP if available else CPU'''
    if torch.cuda.is_available():
        return torch.device('cuda')
    else:
        return torch.device('cpu')
    
def to_device(data,device):
    '''move tensor(s) to choosen device'''
    if isinstance(data,(list,tuple)):
        return [to_device(x,device) for x in data]
    return data.to(device,non_blocking=True)

```

```python
class DeviceDataLoader():
    '''wrap a dataloader to move data to a device'''
    def __init__(self,dl,device):
        self.dl = dl
        self.device = device
        
    def __iter__(self):
        '''yield a batch of data after moving it to device'''
        for b in self.dl:
            yield to_device(b,self.device)
            
    def __len__(self):
        '''number of batches'''
        return len(self.dl)
```
