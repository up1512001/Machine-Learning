```python
import numpy as np

def split_indices(n,val_pct=0.1,seed=99):
    # determine size of validation set
    n_val = int(val_pct*n)
    # set random seed for reproducibility
    np.random.seed(seed)
    # create random permutation of 0 to n-1
    idxs = np.random.permutation(n)
    # pick last n_val indices for validation set
    return idxs[n_val:],idxs[:n_val]
```

```python
def get_default_device():
    '''pick GUP if available else CPU'''
    if torch.cuda.is_available():
        return torch.device('cuda')
    else:
        return torch.device('cpu')
    
def to_device(data,device):
    '''move tensor(s) to choosen device'''
    if isinstance(data,(list,tuple)):
        return [to_device(x,device) for x in data]
    return data.to(device,non_blocking=True)

```

```python
class DeviceDataLoader():
    '''wrap a dataloader to move data to a device'''
    def __init__(self,dl,device):
        self.dl = dl
        self.device = device
        
    def __iter__(self):
        '''yield a batch of data after moving it to device'''
        for b in self.dl:
            yield to_device(b,self.device)
            
    def __len__(self):
        '''number of batches'''
        return len(self.dl)
```

## Training Model 

```python
def loss_batch(model,loss_func,xb,yb,opt=None,metric=None):
    # generates prediction
    preds = model(xb)
    # calculates loss
    loss = loss_func(preds,yb)
    
    if opt is not None:
        # computes gradients
        loss.backward()
        # update parameter 
        opt.step()
        # reset gradients
        opt.zero_grad()
        
    metric_result = None
    
    if metric is not None:
        # computs result metric
        metric_result = metric(preds,yb)
        
    return loss.item(),len(xb),metric_result
```

```python
def evaluate(model,loss_fn,valid_dl,metric=None):
    with torch.no_grad():
        # pass each batch through the model
        results = [loss_batch(model,loss_fn,xb,yb,metric=metric)for xb,yb in valid_dl]
        
        # separate losses , counts , and metric
        losses , nums , metrics = zip(*results)
        
        # total size of dataset
        total = np.sum(nums)
        
        # average loss across batches
        avg_loss = np.sum(np.multiply(losses,nums))/total
        
        avg_metric = None
        if metric is not None:
            # average of metric accross batches
            avg_metric = np.sum(np.multiply(metrics,nums))/total
            
    return avg_loss,total,avg_metric            
```

```python
def fit(epochs,model,loss_fn,train_dl,valid_dl,opt_fn=None,lr=None,metric=None):
    train_losses,val_losses,val_metrics = [],[],[]
    
    # instatiate optimizer
    if opt_fn is None: opt_fn=torch.optim.SGD
    opt = opt_fn(model.parameters(),lr=lr)
    
    for epoch in range(epochs):
        # training
        model.train()
        for xb,yb in train_dl:
            train_loss,_,_ = loss_batch(model,loss_fn,xb,yb,opt)
            
        # evaluation
        model.eval()
        result = evaluate(model,loss_fn,valid_dl,metric)
        val_loss,total,val_metric = result
        
        # record loss & metric
        train_losses.append(train_loss)
        val_losses.append(val_loss)
        val_metrics.append(val_metric)
        
        # print process
        if metric is None:
            print('Epoch [{}/{}] , train_loss : {:.4f} , val_loss : {:.4f}'.format(epoch+1,epochs,train_loss,val_loss))
        else:
            print('Epoch [{}/{}] , train_loss : {:.4f} , val_loss : {:.4f} , val_{} : {:.4f}'.format(epoch+1,epochs,train_loss,
                                                                                                     val_loss,
                                                                                                     metric.__name__,
                                                                                                     val_metric))
            
    return train_losses,val_losses,val_metrics
```

```python
def accuracy(outputs,labels):
    _,preds = torch.max(outputs,dim=1)
    return torch.sum(preds==labels).item()/len(preds)
```

```python
val_loss,_,val_acc = evaluate(model,F.cross_entropy,valid_dl,metric=accuracy)
print('Loss : {:.4f} , Accuracy : {:.4f}'.format(val_loss,val_acc))
```
