```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pd.pandas.set_option('display.max_columns',None)
```

```python
dataset = pd.read_csv('/content/sample_data/Advanced House price prediction/train.csv')
print(dataset.shape)
```

```python
dataset.head()
```

In Data analysis we will analysis to find out below things
1. Missing vales
2. all the numerical variables
3. distribution of numerical variables
4. categorical variables
5. cardinaity of categorical variables
6. outliers
7. relationship between independent and dependent features

```python
# Missing Values

features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum() > 1 ]

for feature in features_with_na:
  print(feature,np.round(dataset[feature].isnull().mean(),4), '% missing values ')
```

```python
# checking for relationship between missing value and sales price

for feature in features_with_na:
  data = dataset.copy()
  
  data[feature] = np.where(data[feature].isnull(),1,0)

  data.groupby(feature)['SalePrice'].median().plot.bar()
  plt.title(feature)
  plt.show()
```

```python
print('Id of Houses :{}'.format(len(dataset.Id)))
```

```python
# Numerical values

numerical_features = [feature for feature in dataset.columns if dataset[feature].dtype!='O']

print('Total Numerical features are {}'.format(len(numerical_features)))

dataset[numerical_features].head()
```

```python
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]

year_feature
```

```python
for feature in year_feature:
  print(feature,dataset[feature].unique())
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title('House Price Vs. Year Sold')
```

```python
year_feature
```

```python
for feature in year_feature:
  if feature!='YrSold':
    data = dataset.copy()
    data[feature] = data['YrSold']-data[feature]

    plt.scatter(data[feature],data['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel('Sale Price')
    plt.title('House Price Vs. {}'.format(feature))
```

```python
discrete_feature = [feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]
print('Total Discrete Features are {}'.format(len(discrete_feature)))
```

```python
discrete_feature
```

```python
dataset[discrete_feature].head()
```

```python
for feature in discrete_feature:
  data = dataset.copy()
  data.groupby(feature)['SalePrice'].median().plot.bar()
  plt.xlabel(feature)
  plt.ylabel('SalePrice')
  plt.title(feature)
  plt.show()
```

```python
# Continuous variables

continuous_feature = [feature for feature in numerical_features if feature not in discrete_feature+year_feature+['Id']]
print('Total Continuous features are {}'.format(len(continuous_feature)))
```

```python
continuous_feature
```

```python
for feature in continuous_feature:
  data = dataset.copy()
  data[feature].hist(bins=25)
  plt.xlabel(feature)
  plt.ylabel('Count')
  plt.title(feature)
  plt.show()
```
