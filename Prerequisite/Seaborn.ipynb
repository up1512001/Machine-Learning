# Seaborn Basic Implementation 


```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
# Correlation with heatmap
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
df.dtypes
```

```python
# Joinplot
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
# pair plot
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='time')
```

```python
df['smoker'].value_counts()
```

```python
sns.pairplot(df,hue_order=['time','sex'])
```

```python
# Dist plot
```

```python
sns.displot(df['tip'])
```

```python
sns.displot(df['tip'],kde=True,bins=10)
```

```python
sns.displot(df['sex'],kde=False,bins=5)
```

```python
# catagorical plots
```

```python
sns.countplot(x='sex',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

```python
sns.countplot(x='smoker',data=df)
```

```python
sns.countplot(x='day',data=df)
```

```python
# bar plot
```

```python
sns.barplot(x='total_bill',y='sex',data=df)
```

```python
df['total_bill'].value_counts()
```

```python
sns.barplot(x='smoker',y='total_bill',data=df)
```

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
# box plot
```

```python
sns.boxplot(x='smoker',y='total_bill',data=df)
```

```python
sns.boxplot(x='total_bill',y='smoker',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df)
```

```python
sns.boxplot(x='sex',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

```python
# violin plot
```

```python
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```

```python
sns.violinplot(x='total_bill',y='sex',data=df,palette='rainbow')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,palette='rainbow',kind='reg')
```

```python
sns.violinplot(x='tip',y='smoker',data=df,palette='rainbow')
```
