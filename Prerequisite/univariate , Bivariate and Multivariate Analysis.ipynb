```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
```

```python
df
```

```python
df.shape
```

## univariate analysis


```python
df_setosa=df.loc[df['species']=='setosa']
```

```python
df_virginica=df.loc[df['species']=='virginica']
```

```python
df_versicolor=df.loc[df['species']=='versicolor']
```

```python
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'r')
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']),'y')
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']),'g')

plt.xlabel('sepal length')
```

```python
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'o')
```

```python
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']),'o')
```

```python
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']),'o')
```

## Bivariate Analysis
