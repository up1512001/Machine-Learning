```python
x = np.arange(0,12*np.pi,0.1)
y = np.sin(x)
plt.title('Sine Wave')
plt.plot(x,y,'r')
```

```python
x = np.arange(0,5*np.pi,0.1)
x
```

```python
y_cos = np.cos(x)
y_sin = np.sin(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin,'g')
plt.title('sin')

plt.subplot(2,1,2)
plt.plot(x,y_cos,'r')
plt.title('cos')
```

```python
x = np.random.randint(1,50,3)
y = np.random.randint(1,50,3)

x2 = [3,6,26]
y2 = [3,6,22]

plt.bar(x,y,align='center')
plt.bar(x2,y2,color='r')
plt.title('bar graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
a = np.random.randint(1,100,10)
plt.hist(a,bins=10)
plt.title('Histogram')
```

```python
# Box plot using matplotlib
data = [np.random.normal(0,std,100)for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True)
```

# Pie Chart


```python
labels = 'python','c','c++','java','golang'
size = [200,105,300,247,40]
colors = ['gold','yellow','yellowgreen','lightcoral','lightskyblue']
explode = (0,0,.11,0,0)

plt.pie(size,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)

plt.axis('equal')
```
